## 6. Memeriksa Kemungkinan Adanya Ketidaksesuaian pada Data

- Kolom 'Order ID' memiliki inisial ID yang merepresentasikan masing-masing region (EMEA, AMER, APJ), kemudian diikuti tahun dan nomor ID. `Contohnya, EMEA-2022-152156`. Untuk memastikan bahwa inputan per kolom Order ID sudah sesuai dengan Region-nya, kita bisa memeriksa kesesuaiannya dengan cara sebagai berikut:

```python
def check_rows_len(s, col_name, desc):
    order_id_match_count = len(df['order_id'].str.startswith(s))
    column_match_count = len(df[col_name] == s)
    
    if order_id_match_count == column_match_count:
        print(f"Panjang baris pada kolom Order ID '{s}' sama dengan panjang baris dari kolom {desc} '{s}'")
    else:
        print(f"Panjang baris tidak sama antara Order ID dan {desc} '{s}'. Cek kembali.")

for region in df['region'].unique():
    check_rows_len(region, 'region', 'Region')

for year in df['order_year'].astype(str).unique():
    check_rows_len(year, 'order_year', 'Order Year')
```

- Memeriksa Apakah Terdapat Spasi di Awal dan/atau akhir pada data dengan value 'string'

```python
def check_extra_spaces(df):
    extra_spaces = {}
    for cols in df.columns:
        if df[cols].dtypes == ['string']:
            # Cek jika ada spasi di awal atau akhir string
            has_extra_space = df[cols].str.startswith(' ') | df[cols].str.endswith(' ')
            if has_extra_space.any():
                extra_spaces[cols] = df[cols][has_extra_space]
    
    if extra_spaces:
        return extra_spaces
    else:
        return "Tidak ada extra spaces atau spasi pada dataset."

check_extra_spaces(df)
```

#### 9.1 Korelasi
Untuk menjawab pertanyaan mengenai faktor-faktor yang mempengaruhi rendahnya profit margin APJ dalam data AWS SaaS, kita perlu menghitung korelasi antar variabel. Dengan mengetahui korelasi, kita dapat mengidentifikasi variabel-variabel yang memiliki hubungan kuat terhadap profit margin. Dalam statistika, kita bisa menggunakan metode Pearson dan Spearman untuk menghitung korelasi. Korelasi adalah ukuran statistik yang menunjukkan seberapa kuat dua variabel saling berhubungan. Nilai korelasi berkisar antara -1 (korelasi negatif sempurna) hingga 1 (korelasi positif sempurna). Sementara korelasi 0 artinya, tidak ada korelasi, dan nilai korelasi mendekati nol berarti korelasi lemah.

Contoh sederhana: Misalnya, terdapat variabel 'sales' yang merupakan data tentang total penjualan produk. Kita ingin mengetahui korelasi antara variabel 'sales' tersebut dengan 'profit'. Setelah dihitung, ternyata nilai korelasi antara kedua variabel tersebut adalah 0.8 (mendekati 1). Artinya, korelasi keduanya kuat dan positif. Positif di sini artinya adalah: **semakin besar nilai 'sales' atau semakin tinggi penjualan produk, maka 'profit' atau keuntungan yang dihasilkan juga semakin besar**. Sebaliknya, jika korelasinya -0.8 artinya korelasi kuat negatif, di mana semakin tinggi penjualan, profit justru menurun, dan sebaliknya. [(Statistics by Jim)](https://statisticsbyjim.com/basics/correlations/)

#### 9.2 Simpson's Paradox
Saat melakukan analisis korelasi antar variabel pada data, kita perlu waspada terhadap fenomena yang disebut Simpson's Paradox. [(If you don't know Simpson's paradox, then you don't understand correlation)](https://www.linkedin.com/pulse/do-you-understand-what-correlation-dont-know-simpsons-zheng-shao/)

**Apa itu Simpson's Paradox?**

Contoh: Misalnya, selain variabel 'sales' dan 'profit', terdapat variabel lain seperti 'product', yakni jenis produk yang dijual. Ketika kita menghitung korelasi antara 'sales' dan 'profit' untuk produk A dan B secara terpisah, hasil korelasinya adalah positif.

--- > `df[df['product_category']=='A'][['sales','profit']].corr(); atau df[df['product_category']=='B'][['sales','profit']].corr()`

Namun, saat kita menghitung korelasi tersebut secara gabungan untuk semua kategori produk, justru korelasinya bisa menjadi negatif (terbalik).

--- > `df[['sales','profit']].corr()`

Ini adalah contoh dari Simpson's Paradox, di mana tren dalam subset data terbalik ketika data digabungkan. Oleh karena itu, dalam menganalisis data AWS SaaS ini, kita perlu mengantisipasi fenomena Simpson's Paradox dengan melakukan analisis korelasi pada subkelompok data (subgroup analysis), selain menghitung korelasi pada data gabungan. Dengan demikian, korelasi dapat dihitung dengan lebih akurat dan tidak hanya berdasarkan data keseluruhan saja. [(Simson's Paradox Explained)](https://en.wikipedia.org/wiki/Simpson%27s_paradox)

## 11. Basket Analisis Produk di APJ

Mengingat bahwa: 
- hasil korelasi `discount vs qty` yang sangat lemah, baik diukur secara keseluruhan maupun per sub data,
- serta tidak ditemukan cukup bukti mengenai perbedaan qty produk pada grup diskon yang berbeda-beda dengan uji ANOVA
bisa disimpulkan dengan kuat, bahwa diskon bukan lah faktor efektif yang meningkatkan penjualan, dan diskon hanya memiliki pengaruh negatif kepada profitabilitas, terlepas dari persentasenya.
- Sehingga, kita perlu beralih ke solusi lainnya, yakni: basket analysis. Dengan basket analysis, kita bisa mengetahui kombinasi produk apa yang paling sering dibeli bersamaan. Dengan begitu, alih-alih memberikan diskon besar-besaran, kita bisa memberikan promosi bundling dan mengetahui strategi product placement yang paling tepat

```python
basket_apj = df_apj.groupby(['order_id', 'product'])['qty'].sum().unstack().reset_index().fillna(0).set_index('order_id')
basket_apj = basket_apj.map(lambda x: True if x > 0 else False)
```

```python
from mlxtend.frequent_patterns import apriori #menggunakan algoritma apriori untuk analisa pola pembelian

frequent_itemsets = apriori(basket_apj, min_support=0.01, use_colnames=True)
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5) #cari kombinasi dengan confidence di atas 50%
```

```python
pd.DataFrame(rules)
```

- Rules di atas menunjukkan bahwa ketika pelanggan membeli produk OneView dan SaaS Connector Pack, terdapat 57.89% kemungkinan bahwa pelanggan akan membeli ContactMatcher (57.89% confidence).
- Kemungkinan pelanggan membeli ContactMatcher setelah membeli OneView dan SaaS Connector Pack adalah 1.85 kali lebih tinggi atau hampir dua kali lipat
- Dengan conviction yang lebih tinggi dari 1, ada indikasi bahwa kehadiran antecedents (OneView dan SaaS Connector Pack) memperkuat kehadiran consequents (ContactMatcher)
